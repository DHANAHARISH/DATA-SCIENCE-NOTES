# Inferential statistics
  population - group that you want to conculsion 
  sample - subset of a population 

    HYPOTHESIS TESTING 
       Process of taking and using sample statistics to make inference(conculsion based on the evidence) for a population            parameter 


    Formulation of hypothesis 
     1.Set up a hypothesis
      *Null hypothesis - there is no significant difference
      *Alternative hypothesis - there is significant difference 
     2.Significant level ( 0.5, 0.1,0.10)
       * is related Ho and denoted by alpha 
       * critical region 
       * confidence level 
     3.Test statistics
       *Suitable for analysis ( level, number , type of data )
       * result score 
           - sample -ztest(formula)-zscore 
           - z table - p.value 
     4.Making decision 
       * reject or accept the null H0 
       * score/ p,value 
       * critical value 
       * significant value 

## z test 

1.Some new survey/research claims that the average age of passengers in Titanic who survived is greater than 28.
                                                                                                                           First, let’s look at the data in hand. As shown below(Graph 1), the population is not normally distributed. So we as per the Central Limit Theorem will take 60 random sampling distribution of 60 sample mean of passengers who survived which will approximate to a normal distribution(Graph 2).



```python
titanic_data=pd.read_csv("D:\dhana_train_data.csv")
```

```python
titanic_data
```

```python
temp=titanic_data[titanic_data['Age'].notna()].Age
sns.distplot(temp)                                  # the population is not normally distributed.
```

```python
survived_data=titanic_data [titanic_data['Survived']==1]
survived_data[survived_data['Age'].notna()].head() #So we as per the Central Limit Theorem will take 60 random sampling  
                                                   #distribution of 60 sample mean
```

```python
sample_list = []
for i in range(60):
    sample_list_temp= np.random.choice(survived_data[survived_data['Age'].notna()].Age,60).mean()
    sample_list.append(sample_list_temp)

sns.distplot(sample_list)
```

```python

mean_h0=28
alpha=0.05
```

```python
from statsmodels.stats.weightstats import ztest 
from statsmodels.stats.weightstats import zconfint
```

```python
z_score,p_value=ztest(sample_list,value=mean_h0,alternative='larger')
```

```python
z_score
```

```python
p_value
```

```python
if p_value<alpha:
    print('reject null h0')
else:
    print('accept null h0')
```

```python
# As per the Z test, we reject H0 and go with the alternate theory which says
# the Average age of survived passengers is > 28. 
```
